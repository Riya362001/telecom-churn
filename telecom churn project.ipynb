```python
# Calculate the average minutes of usage (MOU) during the good phase (months 6 and 7)
data['avg_mou_good'] = (data['total_ic_mou_6'] + data['total_og_mou_6'] + data['total_ic_mou_7'] + data['total_og_mou_7']) / 2

# Calculate the total minutes of usage in the action phase (month 8)
data['total_mou_8'] = data['total_ic_mou_8'] + data['total_og_mou_8']

# Compare the MOU in the action phase (month 8) with the average MOU in the good phase
data['decrease_mou_action'] = data['total_mou_8'] < data['avg_mou_good']

# Convert the boolean values to 1 (decrease) and 0 (no decrease)
data['decrease_mou_action'] = data['decrease_mou_action'].astype(int)

# Display the new column
data[['avg_mou_good', 'total_mou_8', 'decrease_mou_action']].head()
```

```python
# Calculate the average recharge count during the good phase (months 6 and 7)
data['avg_rech_num_good'] = (data['total_rech_num_6'] + data['total_rech_num_7']) / 2

# Compare the recharge count in the action phase (month 8) with the average in the good phase
data['decrease_rech_num_action'] = data['total_rech_num_8'] < data['avg_rech_num_good']

# Convert the boolean values to 1 (decrease) and 0 (no decrease)
data['decrease_rech_num_action'] = data['decrease_rech_num_action'].astype(int)

# Display the new column
data[['avg_rech_num_good', 'total_rech_num_8', 'decrease_rech_num_action']].head()
```

```python
# Calculate the average recharge amount during the good phase (months 6 and 7)
data['avg_rech_amt_good'] = (data['total_rech_amt_6'] + data['total_rech_amt_7']) / 2

# Compare the recharge amount in the action phase (month 8) with the average in the good phase
data['decrease_rech_amt_action'] = data['total_rech_amt_8'] < data['avg_rech_amt_good']

# Convert the boolean values to 1 (decrease) and 0 (no decrease)
data['decrease_rech_amt_action'] = data['decrease_rech_amt_action'].astype(int)

# Display the new column
data[['avg_rech_amt_good', 'total_rech_amt_8', 'decrease_rech_amt_action']].head()
```

```python
# Calculate the average ARPU during the good phase (months 6 and 7)
data['avg_arpu_good'] = (data['arpu_6'] + data['arpu_7']) / 2

# Compare the ARPU in the action phase (month 8) with the average in the good phase
data['decrease_arpu_action'] = data['arpu_8'] < data['avg_arpu_good']

# Convert the boolean values to 1 (decrease) and 0 (no decrease)
data['decrease_arpu_action'] = data['decrease_arpu_action'].astype(int)

# Display the new column
data[['avg_arpu_good', 'arpu_8', 'decrease_arpu_action']].head()
```

```python
# Calculate the average VBC during the good phase (months 6 and 7)
data['avg_vbc_good'] = (data['jun_vbc_3g'] + data['jul_vbc_3g']) / 2

# Compare the VBC in the action phase (month 8) with the average in the good phase
data['decrease_vbc_action'] = data['aug_vbc_3g'] < data['avg_vbc_good']

# Convert the boolean values to 1 (decrease) and 0 (no decrease)
data['decrease_vbc_action'] = data['decrease_vbc_action'].astype(int)

# Display the new column
data[['avg_vbc_good', 'aug_vbc_3g', 'decrease_vbc_action']].head()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import roc_curve, roc_auc_score
cutoff_df = pd.DataFrame(columns=['probability', 'accuracy', 'sensitivity', 'specificity'])

for i in prob_cutoff:
    cm1 = confusion_matrix(y_train_pred_final['churn'], y_train_pred_final[i] )
    total1=sum(sum(cm1))
    accuracy = (cm1[0,0]+cm1[1,1])/total1
    
    speci = cm1[0,0]/(cm1[0,0]+cm1[0,1])
    sensi = cm1[1,1]/(cm1[1,0]+cm1[1,1])
    cutoff_df.loc[i] =[ i ,accuracy,sensi,speci]
print(cutoff_df)
    
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
# Accuracy
print("Accuracy:-",accuracy_score(y_train_pred_final['churn'], y_train_pred_final['predicted']))

# Sensitivity
print("Sensitivity:-",TP / float(TP+FN))

# Specificity
print("Specificity:-", TN / float(TN+FP))
```

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```
